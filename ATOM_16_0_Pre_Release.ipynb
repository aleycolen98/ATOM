```python
def menu_principal():    
    
    from tkinter import ttk
    import tkinter.messagebox
    from PIL import ImageTk, Image
    import time
    from win10toast import ToastNotifier

    root = Tk()
    root.title("ATOM Alpha 16.0 Pre-Release")
    root.geometry("670x300")
    root.config(bg="#fafffa") #Asigna color al frame
    root.wm_iconbitmap('bio.ico')
    root.iconbitmap('bio.ico')
    
    #Posiciona la ventana en el centro
    ancho_ventana = 670
    alto_ventana = 300
    x_ventana = root.winfo_screenwidth() // 2 - ancho_ventana // 2
    y_ventana = root.winfo_screenheight() // 2 - alto_ventana // 2
    posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
    root.geometry(posicion)
    root.resizable (0,0) #Bloquea la ventana para aumento de tamaño
    
    #Imagen de fondo
    img_23=PhotoImage(file="Porta.png")
    fondo_imagen=Label(root,bg="#fafffa", image=img_23)
    fondo_imagen.place(x=-100, y=0)
    
    #Barra Superior
    barraMenu=Menu(root)
    root.config(menu=barraMenu)
    archivoMenu=Menu(barraMenu)
    

    barraMenu.add_command(label="Manual de usuario", command = open_manual)
    barraMenu.add_command(label="Acerca de", command = show_info)
    barraMenu.add_command(label="Código fuente", command =  open_link_code)
    
    #Titulo Principal
    tituloVentana=Label(root, text="BioAtom",fg="#122405", bg="#fafffa",font=("Lucida Bright", 30, "bold"))
    tituloVentana.place(x=410, y=30)
    
    
  
    def data_entrada():
        import tkinter as Tk
        btn.config(state='disable')
        from tkinter import ttk 
        window_1=Tk.Toplevel(root) #Creamos Raíz
        window_1.title('Entrada de datos')
        window_1.iconbitmap('bio.ico')
        import tkinter.messagebox
        import Periodica as E

        ventanaEntrada=Frame(window_1, width=1100, height=600, bg="#e9f8dd") #Creamos Frame Principal
        ventanaEntrada.pack()

        #Posiciona la ventana en el centro
        ancho_ventana = 1100
        alto_ventana = 600
        x_ventana = window_1.winfo_screenwidth() // 2 - ancho_ventana // 2
        y_ventana = window_1.winfo_screenheight() // 2 - alto_ventana // 2
        posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
        window_1.geometry(posicion)
        window_1.resizable (0,0) #Bloquea la ventana para evitar aumento de tamaño
        
        def on_close():  
                  
            window_1.destroy()  # Destruimos la ventana secundaria
            btn.config(state='normal')  # habilitamos el botón
        
       
        "-------------------------------------------------------------------------------------------------------------------------"
                                                                 #TITULOS
        "-------------------------------------------------------------------------------------------------------------------------"
        #PRIMERA FILA
        tagSysName=Label(ventanaEntrada, text="Nombre del sistema : ", bg="#e9f8dd")
        tagSysName.place(x=40,y=65)

        tagSys=Label(ventanaEntrada, text="Etiqueta del sistema : ", bg="#e9f8dd")
        tagSys.place(x=220,y=65)

        #SEGUNDA FILA
        tagNAt=Label(ventanaEntrada, text="Número de átomos : ", bg="#e9f8dd")
        tagNAt.place(x=40,y=125)

        tagNEs=Label(ventanaEntrada, text="Número de especies : ", bg="#e9f8dd")
        tagNEs.place(x=220,y=125)

        tagPRe=Label(ventanaEntrada, text="Parámetro de red : ", bg="#e9f8dd")
        tagPRe.place(x=400,y=125)

        tagCAb=Label(ventanaEntrada, text="Cálculo de bandas : ", bg="#e9f8dd")
        tagCAb.place(x=580,y=125)

        uniPRe=Label(ventanaEntrada, text="Ang", bg="#e9f8dd")
        uniPRe.place(x=527 ,y=150)


        #TERCER FILA

        tagENc=Label(ventanaEntrada, text="Energía de corte : ", bg="#e9f8dd")
        tagENc.place(x=40,y=185)

        uniENc=Label(ventanaEntrada, text="Ry", bg="#e9f8dd")
        uniENc.place(x=167,y=210)

        tagMAs=Label(ventanaEntrada, text="MaxSCFIterations : ", bg="#e9f8dd")
        tagMAs.place(x=220,y=185)

        tagDMTol=Label(ventanaEntrada, text="DM. Tolerance : ", bg="#e9f8dd")
        tagDMTol.place(x=400,y=185)

        tagMDNum=Label(ventanaEntrada, text="MD. NumCGsteps : ", bg="#e9f8dd")
        tagMDNum.place(x=580,y=185)
        #CUARTA FILA

        tagMDMax=Label(ventanaEntrada, text="MD. MaxForceTol : ", bg="#e9f8dd")
        tagMDMax.place(x=40,y=245)

        tagSPol=Label(ventanaEntrada, text="SpinPolarized : ", bg="#e9f8dd")
        tagSPol.place(x=220,y=245)

        tagMDVar=Label(ventanaEntrada, text="MD. VariableCell : ", bg="#e9f8dd")
        tagMDVar.place(x=400,y=245)

        tagConCal=Label(ventanaEntrada, text="¿Continuación de otro cálculo? ", bg="#e9f8dd")
        tagConCal.place(x=580,y=245)

        uniMDMax=Label(ventanaEntrada, text="eV/Ang", bg="#e9f8dd")
        uniMDMax.place(x=165 ,y=270)

        #QUINTA FILA

        tagForCoor=Label(ventanaEntrada, text="Formato de coordenadas átomicas : ", bg="#e9f8dd")
        tagForCoor.place(x=40,y=305)

        #SEXTA FILA

        tagVectorBlock=Label(ventanaEntrada, text="Bloque de Vectores(3x3): ", bg="#e9f8dd")
        tagVectorBlock.place(x=866, y=187)

        tagKpointsB=Label(ventanaEntrada, text="Puntos K(3x4): ", bg="#e9f8dd")
        tagKpointsB.place(x=866, y=307)

        tagCoordAtoFormat=Label(ventanaEntrada, text="Opciones Avanzadas ", bg="#e9f8dd")
        tagCoordAtoFormat.place(x=40, y=393)

        #OPCIONES AVANZADAS
        #ETIQUETAS
        tagPAOBaSy=Label(ventanaEntrada, text="PAO.BasisSize: ", bg="#e9f8dd")
        tagPAOBaSy.place(x=40, y=425)

        tagPAOEnSh=Label(ventanaEntrada, text="PAO.EnergyShift: ", bg="#e9f8dd")
        tagPAOEnSh.place(x=220, y=425)

        tagXCfunc=Label(ventanaEntrada, text="XC.functional: ", bg="#e9f8dd")
        tagXCfunc.place(x=400, y=425)

        tagXCaut=Label(ventanaEntrada, text="XC.autors: ", bg="#e9f8dd")
        tagXCaut.place(x=580, y=425)

        #Etiquetas para unidades
        tagUniEnSh=Label(ventanaEntrada, text="eV", bg="#e9f8dd")
        tagUniEnSh.place(x=350,y=450)


        "-------------------------------------------------------------------------------------------------------------------------"
                                                             #CAJAS DE TEXTO
        "-------------------------------------------------------------------------------------------------------------------------"

        class ToolTip(object):

            def __init__(self, widget):
                self.widget = widget
                self.tipwindow = None
                self.id = None
                self.x = self.y = 0

            def showtip(self, text):
                "Display text in tooltip window"
                self.text = text
                if self.tipwindow or not self.text:
                    return
                x, y, cx, cy = self.widget.bbox("insert")
                x = x + self.widget.winfo_rootx() + 57
                y = y + cy + self.widget.winfo_rooty() +27
                self.tipwindow = tw = Toplevel(self.widget)
                tw.wm_overrideredirect(1)
                tw.wm_geometry("+%d+%d" % (x, y))
                label = Label(tw, text=self.text, justify=LEFT,
                              background="#f9ff9e", relief=SOLID, borderwidth=1,
                              font=("tahoma", "8", "normal"))
                label.pack(ipadx=1)
            def hidetip(self):
                tw = self.tipwindow
                self.tipwindow = None
                if tw:
                    tw.destroy()

        def CreateToolTip(widget, text):
            toolTip = ToolTip(widget)
            def enter(event):
                toolTip.showtip(text)
                widget.config(background="#cee7b6", foreground= "black")
            def leave(event):
                toolTip.hidetip()
                widget.config(background="white", foreground= "black")
            widget.bind('<Enter>', enter)
            widget.bind('<Leave>', leave)

        #PRIMER FILA
        txBoxSys=Entry(ventanaEntrada)
        txBoxSys.place(x=43,y=90)
        #Configuración de evento
        def on_enter1(e):
           txBoxSys.config(background="#cee7b6", foreground= "black")
        def on_leave1(e):
           txBoxSys.config(background="white", foreground= "black")
        txBoxSys.bind('<Enter>', on_enter1)
        txBoxSys.bind('<Leave>', on_leave1)
        CreateToolTip(txBoxSys, text = 'Nombre descriptivo. Establece el nombre del FDF.')



        txBoxSysName=Entry(ventanaEntrada)
        txBoxSysName.place(x=223,y=90)
        #Configuración de evento
        def on_enter2(e):
           txBoxSysName.config(background="#cee7b6", foreground= "black")
        def on_leave2(e):
           txBoxSysName.config(background="white", foreground= "black")
        txBoxSysName.bind('<Enter>', on_enter2)
        txBoxSysName.bind('<Leave>', on_leave2)
        CreateToolTip(txBoxSysName, text="Establece el nombre para\nlos archivos de salida.")

        #SEGUNDA FILA
        txBoxNAt=Entry(ventanaEntrada)
        txBoxNAt.place(x=43,y=150)
        CreateToolTip(txBoxNAt, text = 'Rango permitido:\n'
                         'De 1 a 1000 átomos.')

        txBoxNEs=Entry(ventanaEntrada)
        txBoxNEs.place(x=223,y=150)
        CreateToolTip(txBoxNEs, text = 'Rango permitido:\n'
                         'De 1 a 15 especies.')

        txBoxPRe=Entry(ventanaEntrada)
        txBoxPRe.place(x=403,y=150)
        #Configuración de evento
        def on_enter5(e):
           txBoxPRe.config(background="#cee7b6", foreground= "black")
        def on_leave5(e):
           txBoxPRe.config(background="white", foreground= "black")
        txBoxPRe.bind('<Enter>', on_enter5)
        txBoxPRe.bind('<Leave>', on_leave5)

       #Matriz de vectores
        txBoxVB1_1=Entry(ventanaEntrada)
        txBoxVB1_1.place(x=870, y=210, width=50)
        txBoxVB1_1.insert(INSERT,"0.00000")

        txBoxVB2_1=Entry(ventanaEntrada)
        txBoxVB2_1.place(x=935, y=210, width=50)
        txBoxVB2_1.insert(INSERT,"0.00000")

        txBoxVB3_1=Entry(ventanaEntrada)
        txBoxVB3_1.place(x=1000, y=210, width=50)
        txBoxVB3_1.insert(INSERT,"0.00000")

        txBoxVB1_2=Entry(ventanaEntrada)
        txBoxVB1_2.place(x=870, y=240, width=50)
        txBoxVB1_2.insert(INSERT,"0.00000")

        txBoxVB2_2=Entry(ventanaEntrada)
        txBoxVB2_2.place(x=935, y=240, width=50)
        txBoxVB2_2.insert(INSERT,"0.00000")

        txBoxVB3_2=Entry(ventanaEntrada)
        txBoxVB3_2.place(x=1000, y=240, width=50)
        txBoxVB3_2.insert(INSERT,"0.00000")

        txBoxVB1_3=Entry(ventanaEntrada)
        txBoxVB1_3.place(x=870, y=270, width=50)
        txBoxVB1_3.insert(INSERT,"0.00000")

        txBoxVB2_3=Entry(ventanaEntrada)
        txBoxVB2_3.place(x=935, y=270, width=50)
        txBoxVB2_3.insert(INSERT,"0.00000")

        txBoxVB3_3=Entry(ventanaEntrada)
        txBoxVB3_3.place(x=1000, y=270, width=50)
        txBoxVB3_3.insert(INSERT,"0.00000")


        #Matriz Puntos K
        txBoxKP1_1=Entry(ventanaEntrada)
        txBoxKP1_1.place(x=875, y=330, width=30)
        txBoxKP1_1.insert(INSERT,"00")

        txBoxKP2_1=Entry(ventanaEntrada)
        txBoxKP2_1.place(x=922, y=330, width=30)
        txBoxKP2_1.insert(INSERT,"00")

        txBoxKP3_1=Entry(ventanaEntrada)
        txBoxKP3_1.place(x=971, y=330, width=30)
        txBoxKP3_1.insert(INSERT,"00")

        txBoxKP4_1=Entry(ventanaEntrada)
        txBoxKP4_1.place(x=1015, y=330, width=30)
        txBoxKP4_1.insert(INSERT,"00.0")
        CreateToolTip(txBoxKP4_1, text="Valores permitidos para\ncuarta columna:\nmenores o iguales a 1")
        ###########

        txBoxKP1_2=Entry(ventanaEntrada)
        txBoxKP1_2.place(x=875, y=360, width=30)
        txBoxKP1_2.insert(INSERT,"00")

        txBoxKP2_2=Entry(ventanaEntrada)
        txBoxKP2_2.place(x=922, y=360, width=30)
        txBoxKP2_2.insert(INSERT,"00")

        txBoxKP3_2=Entry(ventanaEntrada)
        txBoxKP3_2.place(x=971, y=360, width=30)
        txBoxKP3_2.insert(INSERT,"00")

        txBoxKP4_2=Entry(ventanaEntrada)
        txBoxKP4_2.place(x=1015, y=360, width=30)
        txBoxKP4_2.insert(INSERT,"00.0")

        ################

        txBoxKP1_3=Entry(ventanaEntrada)
        txBoxKP1_3.place(x=875, y=390, width=30)
        txBoxKP1_3.insert(INSERT,"00")

        txBoxKP2_3=Entry(ventanaEntrada)
        txBoxKP2_3.place(x=922, y=390, width=30)
        txBoxKP2_3.insert(INSERT,"00")

        txBoxKP3_3=Entry(ventanaEntrada)
        txBoxKP3_3.place(x=971, y=390, width=30)
        txBoxKP3_3.insert(INSERT,"00")

        txBoxKP4_3=Entry(ventanaEntrada)
        txBoxKP4_3.place(x=1015, y=390, width=30)
        txBoxKP4_3.insert(INSERT,"00.0")

        #TERCER FILA

        txBoxENc=Entry(ventanaEntrada)
        txBoxENc.place(x=43,y=210)
        CreateToolTip(txBoxENc, text = 'Valor máximo recomendado:\n'
                         '700 Ry.')

        txBoxMAs=Entry(ventanaEntrada)
        txBoxMAs.place(x=223,y=210)
        CreateToolTip(txBoxMAs, text = 'Valor máximo recomendado:\n'
                         '200.')

        txBoxDMTol=Entry(ventanaEntrada)
        txBoxDMTol.place(x=403,y=210)
        CreateToolTip(txBoxDMTol, text = 'Valor máximo recomendado:\n'
                         '1.d-4.')

        txBoxMDNum=Entry(ventanaEntrada)
        txBoxMDNum.place(x=583,y=210)
        CreateToolTip(txBoxMDNum, text = 'Si el valor es 0 (cero):\n'
                         'No se realizará optimización geométrica.')

        #CUARTA FILA
        txBoxMDMax=Entry(ventanaEntrada)
        txBoxMDMax.place(x=43,y=270)
        CreateToolTip(txBoxMDMax, text = 'Rango permitido:\n'
                         'De 0.05 a 0.1.')

        #QUINTA FILA

        txBoxForCoor=Entry(ventanaEntrada)
        txBoxForCoor.place(x=43,y=330)
        txBoxForCoor.insert(INSERT,"Fractional")
        #Configuración de evento
        def on_enter11(e):
           txBoxForCoor.config(background="#cee7b6", foreground= "black")
        def on_leave11(e):
           txBoxForCoor.config(background="white", foreground= "black")
        txBoxForCoor.bind('<Enter>', on_enter11)
        txBoxForCoor.bind('<Leave>', on_leave11)
        CreateToolTip(txBoxForCoor, text = 'Formato de las coordenadas atómicas en la entrada.\n'
                                            'Bohr = Coordenadas cartesianas en bohrs.\n'
                                             'Ang = Coordenadas cartesianas en angstroms.\n'
                                             'ScaledCartesian = Coordenadas cartesianas, unidades de la Lattice Constant.\n'
                                             'Fractional = Hace referencia a Lattice Vectors.')

        #TEXTBOX Opciones Avanzadas
        txBoxPAOBaSy=Entry(ventanaEntrada)
        txBoxPAOBaSy.place(x=43, y=450)
        txBoxPAOBaSy.insert(END, 'DZP')
        #Configuración de evento
        def on_enter12(e):
           txBoxPAOBaSy.config(background="#cee7b6", foreground= "black")
        def on_leave12(e):
           txBoxPAOBaSy.config(background="white", foreground= "black")
        txBoxPAOBaSy.bind('<Enter>', on_enter12)
        txBoxPAOBaSy.bind('<Leave>', on_leave12)


        txBoxPAOEnSh=Entry(ventanaEntrada)
        txBoxPAOEnSh.place(x=223, y=450)
        txBoxPAOEnSh.insert(END, '0.02')
        #Configuración de evento
        def on_enter13(e):
           txBoxPAOEnSh.config(background="#cee7b6", foreground= "black")
        def on_leave13(e):
           txBoxPAOEnSh.config(background="white", foreground= "black")
        txBoxPAOEnSh.bind('<Enter>', on_enter13)
        txBoxPAOEnSh.bind('<Leave>', on_leave13)

        txBoxXCfunc=Entry(ventanaEntrada)
        txBoxXCfunc.place(x=403, y=450)
        txBoxXCfunc.insert(END, 'GGA')
        #Configuración de evento
        def on_enter14(e):
           txBoxXCfunc.config(background="#cee7b6", foreground= "black")
        def on_leave14(e):
           txBoxXCfunc.config(background="white", foreground= "black")
        txBoxXCfunc.bind('<Enter>', on_enter14)
        txBoxXCfunc.bind('<Leave>', on_leave14)
        CreateToolTip(txBoxXCfunc, text = 'LDA = Local Density Approximation\n'
                         'GGA ≡ Generalized Gradient Approximation')


        txBoxXCaut=Entry(ventanaEntrada)
        txBoxXCaut.place(x=583, y=450)
        txBoxXCaut.insert(END, 'PBE')
        #Configuración de evento
        def on_enter15(e):
           txBoxXCaut.config(background="#cee7b6", foreground= "black")
        def on_leave15(e):
           txBoxXCaut.config(background="white", foreground= "black")
        txBoxXCaut.bind('<Enter>', on_enter15)
        txBoxXCaut.bind('<Leave>', on_leave15)
        CreateToolTip(txBoxXCaut, text = 'PBE = Perdew-Burke-Ernzerhof\n'
                                        'CA = Ceperley-Alder\n'
                                        'PZ = Perdew-Zunger\n'
                                        'PW92 = Perdew-Wang-92\n'
                                        'WC = Wu-Cohen\n'
                                        'BLYP = Becke-Lee-Yang-Parr')

        "-------------------------------------------------------------------------------------------------------------------------"
                                                         #SLIDE BOX
        "-------------------------------------------------------------------------------------------------------------------------"

        ComBoxCalBanda=ttk.Combobox(ventanaEntrada, values=["SÍ", "NO"], width=17, height=2, state="readonly")
        ComBoxCalBanda.place(x=583, y=150)
        ComBoxCalBanda.set('NO')
        
        
        ComBoxSPol=ttk.Combobox(ventanaEntrada, values=["TRUE", "FALSE"], width=17, height=2, state="readonly")
        ComBoxSPol.place(x=223, y=270)
        ComBoxSPol.set("FALSE")
        
        ComBoxMDVal=ttk.Combobox(ventanaEntrada, values=["TRUE", "FALSE"], width=17, height=2, state="readonly")
        ComBoxMDVal.place(x=403, y=270)
        ComBoxMDVal.set("FALSE")
        
        ComBoxConCal=ttk.Combobox(ventanaEntrada, values=["SÍ", "NO"], width=17, height=2, state="readonly")
        ComBoxConCal.place(x=583, y=270)
        ComBoxConCal.set("NO")


        "-------------------------------------------------------------------------------------------------------------------------"
                                                            #GENERAR FDF
        "-------------------------------------------------------------------------------------------------------------------------"

        def GENERATE_FDF():

            #Declaramos las variables que guardaran el valor "extraido de las ENTRY"
            vsysname = txBoxSys.get()
            vsyslab = txBoxSysName.get()
            vnumatoms = int(txBoxNAt.get()) 
            vnumofspe = int(txBoxNEs.get())
            vlatticons = txBoxPRe.get() 
            vcalband = ComBoxCalBanda.get()
            venercut = int(txBoxENc.get())
            vmaxscfi = int(txBoxMAs.get())
            vdmtole = txBoxDMTol.get()
            vnumgcste = txBoxMDNum.get()
            vmdmaxftool = float(txBoxMDMax.get())
            vspinpo = ComBoxSPol.get()
            vvarcell = ComBoxMDVal.get()
            vcontical = ComBoxConCal.get()
            vatcoorformat = txBoxForCoor.get()
            vpaobasy = txBoxPAOBaSy.get()
            vpaoensh = txBoxPAOEnSh.get()
            vxcfunc = txBoxXCfunc.get()
            vxcaut = txBoxXCaut.get()
            
            #variables para la matriz de vectores
            vvecat1_1 = txBoxVB1_1.get()
            vvecat2_1 = txBoxVB2_1.get()
            vvecat3_1 = txBoxVB3_1.get()

            vvecat1_2 = txBoxVB1_2.get()
            vvecat2_2 = txBoxVB2_2.get()
            vvecat3_2 = txBoxVB3_2.get()

            vvecat1_3 = txBoxVB1_3.get()
            vvecat2_3 = txBoxVB2_3.get()
            vvecat3_3 = txBoxVB3_3.get()


            #Variables para la matriz de puntos k

            vmatkpoin1_1 = txBoxKP1_1.get()
            vmatkpoin2_1 = txBoxKP2_1.get()
            vmatkpoin3_1 = txBoxKP3_1.get()
            vmatkpoin4_1 = float(txBoxKP4_1.get())

            vmatkpoin1_2 = txBoxKP1_2.get()
            vmatkpoin2_2 = txBoxKP2_2.get()
            vmatkpoin3_2 = txBoxKP3_2.get()
            vmatkpoin4_2 = float(txBoxKP4_2.get())

            vmatkpoin1_3 = txBoxKP1_3.get()
            vmatkpoin2_3 = txBoxKP2_3.get()
            vmatkpoin3_3 = txBoxKP3_3.get()
            vmatkpoin4_3 = float(txBoxKP4_3.get())


            #Comprobación de parametros

            bandera = 0

            if vnumatoms <1 or vnumatoms >10000 :
                bandera = 1
                error1 = "Valor fuera del límite (Debe ser entre 1 y 10000)"
            else:
                error1 = "Valor correcto"

            if vnumofspe < 1 or vnumofspe > 15 :
                bandera = 1
                error2 = "Valor fuera del límite (Debe ser entre 1 y 15)"
            else:
                error2 = "Valor correcto"

            #Puntos K
            if vmatkpoin4_1 < 0 or vmatkpoin4_1 > 1 :
                bandera = 1
                error6 = "Valor fuera de límite (Debe ser menor o igual a 1)"
            else:
                error6 = "Valor Correcto"

            if vmatkpoin4_2 < 0 or vmatkpoin4_2 > 1 :
                bandera = 1
                error7 = "Valor fuera de límite (Debe ser menor o igual a 1)"
            else:
                error7 = "Valor Correcto"

            if vmatkpoin4_3 < 0 or vmatkpoin4_3 > 1 :
                bandera = 1
                error8 = "Valor fuera de límite (Debe ser menor o igual a 1)"
            else:
                error8 = "Valor Correcto"

            if vmdmaxftool < 0.05 or vmdmaxftool > 0.1 :
                bandera = 1
                error3 = "Valor fuera del límite (Debe ser entre 0.05 y 0.1)"
            else:
                error3 = "Valor correcto"


            #Comprobacion calculo de bandas
            if vcalband == "SÍ":
                vcalband = "\nWriteBands    true\n%block BandLines\n1   1.000  1.000  1.000  L\n20  0.000  0.000  0.000  \Gamma\n25  2.000  0.000  0.000  X\n%endblock BandLines"
            else:
                vcalband = ""

            #Comprobacion SpinPolarized

            if vspinpo == "TRUE":
                vspinpo = "SpinPolarized        true"
            else:
                vspinpo = ""

            #Comprobacion MD.VariableCell

            if vvarcell == "TRUE":
                vvarcell = "MD.VariableCell       true"
            else:
                vvarcell = ""

            #Comprobacion continuacion de otro calculo

            if vcontical == "SÍ":
                vcontical = "UseSaveData      true"
                
            else:
                vcontical = ""
            

            #Comprobacion de energía de corte

            if venercut >700:
                error4 = "Tomar en cuenta consideraciones del equipo (ejem. memoria RAM para definir valor mín. y máx.)"
            else:
                error4= "Valor correcto"


            #Comprobación de MaxSCF

            if vmaxscfi>200:
                error5 = "Con un valor mayor a 200 el algoritmo genera inestabilidad (errores)"
            else:
                error5 = "Valor correcto"

            #Parte de la creacion del fdf

            if bandera == 0 :    
                coordenadas=open("archivocoordenadas.txt","r")
                registro=open("Registro.txt","r")
                default_fdf_doc = open('%s.fdf'%(vsysname),'w')
                default_fdf_doc.write("SystemName          %s\n"%(vsysname))
                default_fdf_doc.write("SystemLabel         %s\n"%(vsyslab)+"\n")
                default_fdf_doc.write("NumberOfAtoms        %s\n"%(vnumatoms))
                default_fdf_doc.write("NumberOfSpecies     %s\n"%(vnumofspe)+"\n")
                default_fdf_doc.write("%block ChemicalSpeciesLabel\n")

                datos_regis=registro.read()
                default_fdf_doc.write(datos_regis)

                default_fdf_doc.write("%endblock ChemicalSpeciesLabel\n\n")
                if txBoxPAOBaSy.index("end") != 0:
                    default_fdf_doc.write("PAO.BasisSize    %s\n"%(vpaobasy))
                
                if txBoxPAOEnSh.index("end") != 0:
                    default_fdf_doc.write("PAO.EnergyShift   %s eV\n\n"%(vpaoensh))
                
                if txBoxXCfunc.index("end") != 0:
                    default_fdf_doc.write("XC.functional %s\n"%(vxcfunc)) 
                
                if txBoxXCaut.index("end") != 0:
                
                    default_fdf_doc.write("XC.authors    %s\n\n"%(vxcaut))

                if txBoxPRe.index("end") != 0:
                    default_fdf_doc.write("LatticeConstant   %s"%(vlatticons)+" Ang\n")

                default_fdf_doc.write("%block LatticeVectors\n")

                default_fdf_doc.write("   %s    %s   %s\n"%(vvecat1_1,vvecat2_1,vvecat3_1))
                default_fdf_doc.write("   %s    %s   %s\n"%(vvecat1_2,vvecat2_2,vvecat3_2))
                default_fdf_doc.write("   %s    %s   %s\n"%(vvecat1_3,vvecat2_3,vvecat3_3))

                default_fdf_doc.write("%endblock LatticeVectors\n\n")
                default_fdf_doc.write("%block kgrid_Monkhorst_Pack\n")

                default_fdf_doc.write("   %s %s %s %s\n"%(vmatkpoin1_1,vmatkpoin2_1,vmatkpoin3_1,vmatkpoin4_1))
                default_fdf_doc.write("   %s %s %s %s\n"%(vmatkpoin1_2,vmatkpoin2_2,vmatkpoin3_2,vmatkpoin4_2))
                default_fdf_doc.write("   %s %s %s %s\n"%(vmatkpoin1_3,vmatkpoin2_3,vmatkpoin3_3,vmatkpoin4_3))

                default_fdf_doc.write("%endblock kgrid_Monkhorst_Pack\n")
                default_fdf_doc.write("%s\n\n"%(vcalband))
                default_fdf_doc.write("MeshCutoff        %s Ry \n\n"%(venercut))
                default_fdf_doc.write("MaxSCFIterations    %s\n"%(vmaxscfi))
                default_fdf_doc.write("DM.MixingWeight      0.1\n")
                default_fdf_doc.write("DM.NumberPulay       4\n")
                
                
                default_fdf_doc.write("DM.Tolerance        %s\n\n"%(vdmtole))
                    
                default_fdf_doc.write("SolutionMethod       diagon\n\n")
                default_fdf_doc.write("MD.TypeOfRun         cg\n")
                default_fdf_doc.write("MD.NumCGsteps       %s\n"%(vnumgcste))
                default_fdf_doc.write("MD.MaxForceTol      %s eV/Ang\n\n"%(vmdmaxftool))
                default_fdf_doc.write("%s\n"%(vspinpo))
                default_fdf_doc.write("%s\n\n"%(vvarcell))
                default_fdf_doc.write("%s\n\n"%(vcontical))
                default_fdf_doc.write("AtomicCoordinatesFormat %s\n"%(vatcoorformat))
                default_fdf_doc.write("%block AtomicCoordinatesAndAtomicSpecies\n")

                datos_coor=coordenadas.read()
                default_fdf_doc.write(datos_coor)

                default_fdf_doc.write("%endblock AtomicCoordinatesAndAtomicSpecies\n")
                default_fdf_doc.close()
                coordenadas.close()


                error_gen = "--- Archivo generado satisfactoriamente ----"
                tkinter.messagebox.showinfo( "Reporte de Errores", "Número de átomos: %s\n"%error1+"Número de especies: %s\n"%error2+"MD.MaxForceTol: %s\n\n"%error3+"Matriz de puntos K\n\nRenglón 1, Columna 4: %s"%error6+"\nRenglón 2, Columna 4: %s"%error7+ "\nRenglón 3, Columna 4: %s"%error8+"\n\n%s"%error_gen+"\n\n\nAdvertencias: \n\nEnergía de Corte: %s"%error4+ "\nMaxSCFIterations: %s"%error5)

            if bandera == 1:

                error_gen = "El archivo no se ha generado debido a errores en los parámetros"
                tkinter.messagebox.showwarning( "Reporte de Errores", "Número de átomos: %s\n"%error1+"Número de especies: %s\n"%error2+"MD.MaxForceTol: %s\n\n\n\n"%error3+"Matriz de puntos K\n\nRenglón 1, Columna 4: %s"%error6+"\nRenglón 2, Columna 4: %s"%error7+"\nRenglón 3, Columna 4: %s"%error8+"\n\n%s"%error_gen+"\n\n\nAdvertencias: \n\nEnergía de Corte: %s"%error4+ "\nMaxSCFIterations: %s"%error5)


        "-------------------------------------------------------------------------------------------------------------------------"
                                                      #MATRIZ DE ESPECIES QUÍMICAS
        "-------------------------------------------------------------------------------------------------------------------------"
        def gen_especies_quimicas():
            valor=int(txBoxNEs.get())
            if valor>15:
                tkinter.messagebox.showerror("Error","Cantidad de especies fuera de rango.")
                return 0
            else:
                
                btnGenSpChee.config(state='disabled')
                w_1_1=Tk.Toplevel(window_1)
                w_1_1.title("Etiqueta de Especies Químicas")
                w_1_1.iconbitmap('bio.ico')
                
                
                #Posiciona la ventana en el centro
                ancho_ventana = 400
                alto_ventana = 200
                x_ventana = w_1_1.winfo_screenwidth() // 2 - ancho_ventana // 2
                y_ventana = w_1_1.winfo_screenheight() // 2 - alto_ventana // 2
                posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
                w_1_1.geometry(posicion)
                w_1_1.resizable (0,0) #Bloquea la ventana para aumento de tamaño
                
                def on_close5():  
                  
                    w_1_1.destroy()  # Destruimos la ventana secundaria
                    btnGenSpChee.config(state='normal')  # habilitamos el botón
                
                miFrame1=Frame(w_1_1, bg="#c2daaf")
                miFrame1.pack(fill=BOTH, expand=1)
                miCanvas=Canvas(miFrame1, bg="#c2daaf")
                miCanvas.pack(side=LEFT, fill=BOTH, expand=1)
                miScrollbar=ttk.Scrollbar(miFrame1, orient=VERTICAL, command=miCanvas.yview)
                miScrollbar.pack(side=RIGHT, fill=Y)
                miCanvas.configure(yscrollcommand=miScrollbar.set)
                miCanvas.bind('<Configure>', lambda e: miCanvas.configure(scrollregion=miCanvas.bbox("all")))
                miFrame2=Frame(miCanvas, bg="#c2daaf")
                miCanvas.create_window((0,0), window=miFrame2, anchor="nw"),
                Elemento=[]
                for x in range(1, valor+1):
                        Entrada=Entry(miFrame2)#textvariable=Elemento1)
                        Entrada.grid(row=x, column=2, padx=10, pady=10)
                        Etiqueta=Label(miFrame2, text=x, bg="#c2daaf")
                        Etiqueta.grid(row=x,column=1,padx=10,pady=10)
                        Elemento.append(Entrada)
                def send_data():

                    newfile=open("Registro.txt","w")
                    x=0
                    end=valor*3
                    for x in range(0,valor):
                        d=str(x+1)
                        Ent1=Elemento[x].get()
                        newfile.write(d+" "+E.Elementos[Ent1]+"\n")
                    newfile.close()
                    newfile2=open("Registro.txt","r")
                    print(newfile2.read())

                cuadroNombre1=Label(miFrame2, text="               ", bg="#c2daaf")
                cuadroNombre1.grid(row=0, column=0, padx=10, pady=5) 
                cuadroNombre2=Label(miFrame2, text="No.", bg="#c2daaf")
                cuadroNombre2.grid(row=0, column=1, padx=10, pady=5) 
                cuadroNombre3=Label(miFrame2, text="Elemento", bg="#c2daaf")
                cuadroNombre3.grid(row=0, column=2, padx=10, pady=5)

                btn1=Button(miFrame2, text="Enviar", command=send_data,width=12 )
                btn1.grid(row=0, column=5, padx=10, pady=10)
                
                w_1_1.protocol("WM_DELETE_WINDOW", on_close5) 
               

        "-------------------------------------------------------------------------------------------------------------------------"
                                                     #MATRIZ DE COORDENADAS ÁTOMICAS
        "-------------------------------------------------------------------------------------------------------------------------"

        def coordenadas():
            valor=int(txBoxNAt.get())
            if valor>10000:
                tkinter.messagebox.showerror("Error","Cantidad de átomos fuera de rango.")
                return 0
            else:
                
                btnGenCoordAtom.config(state='disabled')
                w_1_2=Tk.Toplevel(window_1)
                w_1_2.title("Coordenadas Atómicas")
                w_1_2.iconbitmap('bio.ico')

                #Posiciona la ventana en el centro
                ancho_ventana = 700
                alto_ventana = 200
                x_ventana = root.winfo_screenwidth() // 2 - ancho_ventana // 2
                y_ventana = root.winfo_screenheight() // 2 - alto_ventana // 2
                posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
                w_1_2.geometry(posicion)
                w_1_2.resizable (0,0) #Bloquea la ventana para aumento de tamaño
                
                def on_close6():  
                  
                    w_1_2.destroy()  # Destruimos la ventana secundaria
                    btnGenCoordAtom.config(state='normal')  # habilitamos el botón
                
                miFrame1=Frame(w_1_2, bg="#c2daaf")
                miFrame1.pack(fill=BOTH, expand=1)
                miCanvas=Canvas(miFrame1, bg="#c2daaf")
                miCanvas.pack(side=LEFT, fill=BOTH, expand=1)
                miScrollbar=ttk.Scrollbar(miFrame1, orient=VERTICAL, command=miCanvas.yview)
                miScrollbar.pack(side=RIGHT, fill=Y)
                miCanvas.configure(yscrollcommand=miScrollbar.set)
                miCanvas.bind('<Configure>', lambda e: miCanvas.configure(scrollregion=miCanvas.bbox("all")))
                miFrame2=Frame(miCanvas, bg="#c2daaf")
                miCanvas.create_window((0,0), window=miFrame2, anchor="nw"),
                Elemento1=[]
                Elemento2=[]
                Elemento3=[]
                Elemento4=[]
                for x in range(1, valor+1):
                        Etiqueta_1=Entry(miFrame2)#textvariable=Elemento1)
                        Etiqueta_1.grid(row=x, column=0, padx=10, pady=10)
                        Etiqueta_1.insert(INSERT,"0.00000")

                        Etiqueta_2=Entry(miFrame2)
                        Etiqueta_2.grid(row=x,column=1,padx=10,pady=10)
                        Etiqueta_2.insert(INSERT,"0.00000")

                        Etiqueta_3=Entry(miFrame2)
                        Etiqueta_3.grid(row=x,column=2,padx=10,pady=10)
                        Etiqueta_3.insert(INSERT,"0.00000")

                        Etiqueta_4=Entry(miFrame2)
                        Etiqueta_4.grid(row=x,column=3,padx=10,pady=10)
                        Etiqueta_4.insert(INSERT,"0")

                        Elemento1.append(Etiqueta_1)
                        Elemento2.append(Etiqueta_2)
                        Elemento3.append(Etiqueta_3)
                        Elemento4.append(Etiqueta_4)
                def send_data():

                    newfile=open("archivocoordenadas.txt","w")
                    x=0
                    end=valor*3
                    for x in range(0,valor):
                        d=str(x+1)
                        Ent1=Elemento1[x].get()
                        Ent2=Elemento2[x].get()
                        Ent3=Elemento3[x].get()
                        Ent4=Elemento4[x].get()
                        a=str(Ent1)
                        b=str(Ent2)
                        c=str(Ent3)
                        e=str(Ent4)
                        newfile.write(a+" "+b+" "+c+" "+e+"\n")
                    newfile.close()
                    newfile2=open("archivocoordenadas.txt","r")
                    print(newfile2.read())

                btn1=Button(miFrame2, text="Enviar", command=send_data,width=10)
                btn1.grid(row=1, column=5, padx=10, pady=10)
                   
                w_1_2.protocol("WM_DELETE_WINDOW", on_close6)    
                
        "-------------------------------------------------------------------------------------------------------------------------"
                                                         #BARRAS LATERALES
        "-------------------------------------------------------------------------------------------------------------------------"   
        matrices_boton=Frame(window_1, width=3, height=78, bg="gray")
        matrices_boton.place(x=840, y=67)

        matrices_etiquetas=Frame(window_1, width=3, height=225, bg="gray")
        matrices_etiquetas.place(x=840, y=191)

        formulario_1=Frame(window_1, width=3, height=287, bg="gray")
        formulario_1.place(x=30, y=67)

        formulario_2=Frame(window_1, width=3, height=80, bg="gray")
        formulario_2.place(x=30, y=395)

        titular=Label(ventanaEntrada, text="Entrada de Datos", bg="#e9f8dd",font=("Gill Sans MT", 17, "bold") )
        titular.place(x=455,y=10)

        "-------------------------------------------------------------------------------------------------------------------------"
                                                         #MENÚ PRINCIPAL
        "-------------------------------------------------------------------------------------------------------------------------"

        #BOTONES AUXILIARES
        btnGenSpChee=Button(ventanaEntrada, text="Generar Etiqueta de Especies Químicas", width=30, height=1, command=gen_especies_quimicas)
        btnGenSpChee.place(x=850, y=70)

        btnGenCoordAtom=Button(ventanaEntrada, text="Ingresar Coordenadas Atómicas", width=30, height=1, command=coordenadas)
        btnGenCoordAtom.place(x=850, y=110)

        #BOTONES DE FINALIZACIÓN
        btnGenFDF=Button(ventanaEntrada, text="Generar FDF", width=17, height=2, bg="#bacea6", command = GENERATE_FDF)
        btnGenFDF.place(x=400, y=510)

        btnVolver=Button(ventanaEntrada, text="Cancelar", width=17, height=2, bg="#bacea6",command = on_close)
        btnVolver.place(x=550, y=510)
        
        window_1.protocol("WM_DELETE_WINDOW", on_close) 
    
    
    #Implementacion del segundo modulo: Conexión a servidor y envio de simulación
        
    def correr_simulacion():
        import tkinter as Tk
        btn2.config(state='disabled')
        
        from tkinter import ttk
        import tkinter.messagebox
        from tkinter import filedialog #Permite abrir el explorador de archivos
        import paramiko
        from scp import SCPClient

        window_2 = Tk.Toplevel(root)
        window_2.title("Correr Simulación")
        window_2.config(bg="#e9f8dd")
        window_2.wm_iconbitmap('bio.ico')
        window_2.iconbitmap('bio.ico')

        #Posiciona la ventana en el centro
        ancho_ventana = 650
        alto_ventana = 370
        x_ventana = window_2.winfo_screenwidth() // 2 - ancho_ventana // 2
        y_ventana = window_2.winfo_screenheight() // 2 - alto_ventana // 2
        posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
        window_2.geometry(posicion)
        window_2.resizable (0,0) #Bloquea la ventana para aumento de tamaño
        
        
        
        def on_close2():  
               
            window_2.destroy()  # Destruimos la ventana secundaria
            btn2.config(state='normal')  # habilitamos el botón

        #conec_status = 1  #variable auxiliar para el estado de la conexion al servidor
        client = paramiko.SSHClient()

        #variable global para el nombre del archivo fdf de la simulacion y para guardar directorio
        nomfdf = ""


        tituloVentana=Label(window_2, text="Conexión a servidor y simulación",bg="#e9f8dd",font=("Gill Sans MT", 17, "bold"))
        tituloVentana.place(x=170, y=30)

        numProce=Label(window_2, text="Cantidad de Procesadores: ", bg="#e9f8dd")
        numProce.place(x=67, y=130)

        defCol=Label(window_2, text="Cola: ", bg="#e9f8dd")
        defCol.place(x=67, y=190)

        deftxIP=Label(window_2, text="IP", bg="#e9f8dd")
        deftxIP.place(x=267, y=130)

        deftxPORT=Label(window_2, text="Puerto", bg="#e9f8dd")
        deftxPORT.place(x=267, y=190)

        deftxUSER=Label(window_2, text="Usuario", bg="#e9f8dd")
        deftxUSER.place(x=457, y=130)

        deftxPASS=Label(window_2, text="Contraseña", bg="#e9f8dd")
        deftxPASS.place(x=457, y=190)

        #Cajas de Texto

        ##entry de procesadores
        txBoxArch=Entry(window_2, state="disable")
        txBoxArch.place(x=70, y=155)


        ##entry cola
        txBoxCol=Entry(window_2, state="disable")
        txBoxCol.place(x=70, y=215)


        ##entry IP
        txBoxIP=Entry(window_2)
        txBoxIP.place(x=270, y=155)
        txBoxIP.insert(END, '148.204.218.97')

        ##entry PUERTO
        txBoxPORT=Entry(window_2)
        txBoxPORT.place(x=270, y=215)
        txBoxPORT.insert(END, '22')

        ##entry USUARIO
        txBoxUSER=Entry(window_2)
        txBoxUSER.place(x=460, y=155)

        ##entry PASSWORD
        txBoxPASS=Entry(window_2)
        txBoxPASS.place(x=460, y=215)
        txBoxPASS.config(show="*")

        #Botones
        txBoxArch2=Entry(window_2)
        txBoxCol2=Entry(window_2)

        btnIngArch1=Button(window_2, text="Subir archivo", width=17, height=1, command = advert_doc)
        btnIngArch1.place(x=70, y=85)



        #Función para establece la conexion con el servidor remoto
        def SERVER_CONNECTION():

            v_ip = txBoxIP.get()
            v_port = txBoxPORT.get()
            v_user = txBoxUSER.get()
            v_pass = txBoxPASS.get()
            #conec_status = 0 

            client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

            try:

                client.connect(v_ip,v_port,v_user,v_pass)
                channel = client.invoke_shell()
                tkinter.messagebox.showinfo( "Conexión Exitosa", "Se ha conectado correctamente al servidor")

                command = "ls"
                stdin, stdout, stderr = client.exec_command(command)
                lines = stdout.readlines()
                print("Directorios\n")
                print(lines)

                txBoxArch2.place(x=70, y=155)
                txBoxArch2.insert(END, '0')

                txBoxCol2.place(x=70, y=215)
                txBoxCol2.insert(END, '0')

                btnIngArch=Button(window_2, text="Subir archivo", width=17, height=1, command = browseFiles)
                btnIngArch.place(x=70, y=85)

                btnDown.config(state = 'normal')
                
                return True

            except paramiko.AuthenticationException:

                tkinter.messagebox.showerror("Error","Conexión fallida, revisa tus credenciales")



        #Abrir y cargar archivo desde el explorador
        def browseFiles():  #Configuración de la dirección de archivo (tipo)
                import os
                nomArchivo = filedialog.askopenfilename(initialdir = "/", 
                                                  title = "Selecciona el archivo", 
                                                  filetypes = (("Archivo FDF", 
                                                                "*.fdf*"), 
                                                               ("all files", 
                                                                "*.*"))) 

                mensaje_archivo.configure(text="Archivo: "+nomArchivo, fg="#17919c", bg="#e9f8dd") #Modifica mensaje de archivo cargado

                scp=SCPClient(client.get_transport())
                scp.put(r"%s"%(nomArchivo),"~")
                nomdic = nomArchivo
                nomfdf = os.path.basename(nomArchivo)
                temp1 = open('temp1','w')
                temp1.write("%s"%nomfdf)
                temp1.close

                tkinter.messagebox.showinfo( "Archivo Subido", "Script subido satisfactoriamente")


        mensaje_archivo=Label(window_2, text="Archivo: No Existente", fg="#c51e5a", bg="#e9f8dd") #Mensaje de archivo cargado
        mensaje_archivo.place(x=267, y=85)

        #Función para mandar a correr la simulacion
        def run_others_commands():

            try:
                print(txBoxArch2.get()) #Procesadores
                print(txBoxCol2.get())  #Cola

                npro = txBoxArch2.get()
                ncola = txBoxCol2.get()

                inputname=open("temp1","r")
                namefdf=inputname.read()
                print(namefdf)

                command4 = "qsub -N ATOMCola -q A%s"%ncola+" -S /bin/bash -V -m aesb  -cwd  -pe mpi %s"%npro+" ./siesta.sh %s"%npro+" %s"%namefdf+" ATOMSalida.out"

                print(command4)
                stdin = client.exec_command(command4)
                
                tkinter.messagebox.showwarning( "Simulación Enviada","Se ha enviado la simulación a la cola.")
            except:

                tkinter.messagebox.showerror( "Error","No se ha establecido conexión con el servidor.")

        #Funcion para descargar el archivo generado al terminar la simulacion de SIESTA
        def download_archive():
            import os

            tiempo_busqueda = time.time() + 60*60
            while time.time() < tiempo_busqueda:
                
                tkinter.messagebox.showinfo("Descarga de archivos","Se notificará cuando se descarguen los archivos\nno cierre la ventana Simulación.")
                scp=SCPClient(client.get_transport())
                scp.get(r"OUTVARS.yml")
                scp.get(r"ATOMSalida.out")
                file_exists = os.path.exists('OUTVARS.yml')
                if file_exists == True:
                    
                    print("file find")
                    toast = ToastNotifier()
                    toast.show_toast("BioAtom","Se ha descargado el archivo OUTVARS.yml y ATOMSalida.out de la última simulación, consúltalo en la carpeta del programa", threaded=True,duration=30,icon_path="bio.ico")
                    tiempo_busqueda = time.time()-60*60
                    btn2.config(state='normal')
                    client.close()
                    window_2.destroy()


        def close_connection():

            client.close()
            window_2.destroy()


        #Boton Inicio
        btnIniSimul=Button(window_2, text="Iniciar", width=15, height=1,bg="#bacea6",command = run_others_commands)
        btnIniSimul.place(x=73, y=270)

        #Boton Conectar
        btnCOSV=Button(window_2, text="Conectar", width=15, height=1, command=SERVER_CONNECTION)
        btnCOSV["activebackground"]="#faf0ff"
        btnCOSV.place(x=437, y=270)

        #Boton Cancelar
        btnVolver=Button(window_2, text="Cancelar",width=15, height=1, command = close_connection)
        btnVolver["activebackground"]="#faf0ff"
        btnVolver.place(x=297, y=270)


        #Se genera un boton para descargar el archivo generado por la simulación
        btnDown=Button(window_2, text="Descargar", width=15, height=1, command=download_archive)
        btnDown["activebackground"]="#faf0ff"
        btnDown.place(x=75, y=300)
        btnDown.config(state = 'disabled')

        window_2.protocol("WM_DELETE_WINDOW", on_close2) 


    def w_resultados():
        import tkinter as Tk
        btn5.config(state='disabled')
        from tkinter import ttk
        import tkinter.messagebox

        window_3=Tk.Toplevel(root)
        window_3.title("Resultados")
        window_3.iconbitmap('bio.ico')

        ventana_resultados=Frame(window_3,width=1100, height=600, bg="#e9f8dd")
        ventana_resultados.pack()

        #Posiciona la ventana en el centro
        ancho_ventana = 800
        alto_ventana = 300
        x_ventana = window_3.winfo_screenwidth() // 2 - ancho_ventana // 2
        y_ventana = window_3.winfo_screenheight() // 2 - alto_ventana // 2
        posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
        window_3.geometry(posicion)
        window_3.resizable (0,0) #Bloquea la ventana para aumento de tamaño
        
        def on_close3():  
                  
                    window_3.destroy()  # Destruimos la ventana secundaria
                    btn5.config(state='normal')  # habilitamos el botón


        """---------------------------------------------------------------------------------------------------------------------"""
                                                          # Numero de atomos
        """---------------------------------------------------------------------------------------------------------------------"""

        archivo=open("ATOMSalida.out", "r") #Abrimos el archivo

        for linea in archivo.readlines(): #Comenzamos a leer línea por línea
            if 'NumberOfAtoms' in linea:
                linea.lstrip() #Elimina caracteres iniciales
                linea.rstrip() #Elimina caracateres finales
                nueva_linea=linea.split() #Guarda en una nueva lista

                valor=int(nueva_linea[1])
                print(valor)

        archivo.close() #Cerramos el archivo

        horario=open("ATOMSalida.out", "r") #Abrimos el archivo
        for linea in horario.readlines(): #Comenzamos a leer línea por línea
            if '>> Start of run:' in linea:
                linea.lstrip() #Elimina caracteres iniciales
                linea.rstrip() #Elimina caracateres finales
                hora=linea.split() #Guarda en una nueva lista

                valor1=hora[4]
                valor2=hora[5]
                print(valor1)
                print(valor2)

        horario.close() #Cerramos el archivo

        nombreSis=open("ATOMSalida.out", "r") #Abrimos el archivo
        for linea in nombreSis.readlines(): #Comenzamos a leer línea por línea
            if 'SystemName' in linea:
                linea.lstrip() #Elimina caracteres iniciales
                linea.rstrip() #Elimina caracateres finales
                nombre=linea.split() #Guarda en una nueva lista

                nomSis=nombre[1]
                print(nomSis)

        nombreSis.close() #Cerramos el archivo



        """---------------------------------------------------------------------------------------------------------------------"""
                                                          #Tabla de atomos 
        """---------------------------------------------------------------------------------------------------------------------"""
        def generar_energias():
            btn6.config(state="disabled")
            win_3_1=Tk.Toplevel(window_3)
            win_3_1.title("Resultados")
            win_3_1.iconbitmap('bio.ico')
            #valor=15

            #Posiciona la ventana en el centro
            ancho_ventana = 560
            alto_ventana = 200
            x_ventana = win_3_1.winfo_screenwidth() // 2 - ancho_ventana // 2
            y_ventana = win_3_1.winfo_screenheight() // 2 - alto_ventana // 2
            posicion = str(ancho_ventana) + "x" + str(alto_ventana) + "+" + str(x_ventana) + "+" + str(y_ventana)
            win_3_1.geometry(posicion)
            win_3_1.config(bg="#c2daaf")
            win_3_1.resizable (0,0) #Bloquea la ventana para aumento de tamaño
            
            def on_close7():  
               
                win_3_1.destroy()  # Destruimos la ventana secundaria
                btn6.config(state='normal')  # habilitamos el botón

            miFrame1=Frame(win_3_1, bg="#c2daaf")
            miFrame1.pack(fill=BOTH, expand=1)
            miCanvas=Canvas(miFrame1, bg="#c2daaf")
            miCanvas.pack(side=LEFT, fill=BOTH, expand=1)
            miScrollbar=ttk.Scrollbar(miFrame1, orient=VERTICAL, command=miCanvas.yview)
            miScrollbar.pack(side=RIGHT, fill=Y)
            miCanvas.configure(yscrollcommand=miScrollbar.set)
            miCanvas.bind('<Configure>', lambda e: miCanvas.configure(scrollregion=miCanvas.bbox("all")))
            miFrame2=Frame(miCanvas, bg="#c2daaf")
            miCanvas.create_window((0,0), window=miFrame2, anchor="nw"),
            Elemento1=[]
            Elemento2=[]
            Elemento3=[]

            tagNombre=Label(miFrame2, text="Nombre de Átomo", bg="#c2daaf")
            tagNombre.grid(row=0,column=0, padx=10, pady=9)
            tagCantidad=Label(miFrame2, text="Cantidad de Átomos", bg="#c2daaf")
            tagCantidad.grid(row=0,column=1, padx=10, pady=9)
            tagEnergia=Label(miFrame2, text="Energía", bg="#c2daaf")
            tagEnergia.grid(row=0, column=2, padx=10, pady=9)


            for x in range(1, valor+1):
                Etiqueta_1=Entry(miFrame2)#textvariable=Elemento1)
                Etiqueta_1.grid(row=x, column=0, padx=10, pady=9)

                Etiqueta_2=Entry(miFrame2)
                Etiqueta_2.grid(row=x,column=1,padx=10,pady=9)

                Etiqueta_3=Entry(miFrame2)
                Etiqueta_3.grid(row=x,column=2,padx=10,pady=9)

                Elemento1.append(Etiqueta_1)
                Elemento2.append(Etiqueta_2)
                Elemento3.append(Etiqueta_3)


            def send_data():

                newfile=open("archivoenergias.txt","w")
                x=0
                for x in range(0,valor):
                    d=str(x+1)
                    Ent1=Elemento1[x].get()
                    Ent2=Elemento2[x].get()
                    Ent3=Elemento3[x].get()
                    a=str(Ent1)
                    b=str(Ent2)
                    c=str(Ent3)

                    newfile.write(a+" "+b+" "+c+"\n")
                newfile.close()

                newfile2=open("archivoenergias.txt","r")
                print(newfile2.read())
                newfile2.close()

                """-----------------------------------------------------------------------------------------------------------------"""
                                                             #Imprimir Resultados Finales
                """-----------------------------------------------------------------------------------------------------------------"""

                y=0
                ener_final=resultado_2
                sum_atom=0
                for y in range(0,valor):
                    ener_final=ener_final-(float(Elemento2[y].get())*float(Elemento3[y].get()))
                    sum_atom=sum_atom+int(Elemento2[y].get())
                ener_final=ener_final/sum_atom
                print("Resultados")
                print(ener_final)
                print(sum_atom)
                print("Energía Final")
                print(ener_final)


                enerTot=Label(ventana_resultados, text="Energía total:", bg="#e9f8dd",font=("Gill Sans MT", 12) )
                enerTot.place(x=250, y=60)

                txBoxEnergíaTot=Entry(ventana_resultados)
                txBoxEnergíaTot.place(x=253, y=90)
                txBoxEnergíaTot.insert(END, resultado_2)

                enerFin=Label(ventana_resultados, text="Energía de cohesión por átomo:", bg="#e9f8dd",font=("Gill Sans MT", 12) )
                enerFin.place(x=250, y=120)

                txBoxEnergíaFin=Entry(ventana_resultados)
                txBoxEnergíaFin.place(x=253, y=150)
                txBoxEnergíaFin.insert(END, ener_final)

                enerFinUnidad=Label(ventana_resultados, text="Ry", bg="#e9f8dd",font=("Gill Sans MT", 12) )
                enerFinUnidad.place(x=380, y=147)

                enerDatos=Label(ventana_resultados, text=nomSis+' | '+valor1+' | '+valor2, bg="#e9f8dd",font=("Gill Sans MT", 12) )
                enerDatos.place(x=500, y=200)
                
                
        
            btn1=Button(miFrame2, text="Completar", command=send_data,width=10)
            btn1.grid(row=0, column=3, padx=10, pady=9)
            
            win_3_1.protocol("WM_DELETE_WINDOW", on_close7)


        """-----------------------------------------------------------------------------------------------------------------"""
                                                          #Energías Finales
        """-----------------------------------------------------------------------------------------------------------------"""


        archivo=open("OUTVARS.yml", "r") #Abrimos el archivo

        for linea in archivo.readlines(): #Comenzamos a leer línea por línea
            if 'Etot:' in linea:
                linea.lstrip() #Elimina caracteres iniciales
                linea.rstrip() #Elimina caracateres finales
                nueva_linea=linea.split() #Guarda en una nueva lista
                e_total=nueva_linea[1] #Recupero valor de Energía Total

                nueva_lista=e_total.split('E') #Sucesión de operaciones para Base10
                valor_1=int(nueva_lista[1])
                resultado_1=pow(10,valor_1)

                valor_2=float(nueva_lista[0])
                resultado_2=resultado_1*valor_2
                print(resultado_2)


        archivo.close() #Cerramos el archivo





        """-----------------------------------------------------------------------------------------------------------------"""
                                                          #Principal
        """-----------------------------------------------------------------------------------------------------------------"""
        primario=Frame(window_3, bg="#242e0f", width=205, height=395)
        primario.place(x=0,y=0)

        btn6=Button(primario, text="Ingresar Energías", command=generar_energias, width=22, height=1)
        btn6.place(x=20, y=110)
        btn3=Button(primario, text="Guardar Resultados",bg="#bacea6",width=22, height=1, command = ERROR_MMS) 
        btn3.place(x=20, y=140)


        titular=Label(ventana_resultados, text="Resultados", bg="#e9f8dd",font=("Gill Sans MT", 17, "bold") )
        titular.place(x=440, y=10)

        window_3.protocol("WM_DELETE_WINDOW", on_close3) 
        
            
    
    #Opciones Principales
    btn = Button(root, text="Entrada", fg="black", font=("Cambria", 12), command = data_entrada)
    btn["height"]=3
    btn["width"]=15
    btn["activebackground"]="#faf0ff"
    btn["fg"]="black"
    btn["bg"]="white"
    btn.place(x=75,y=120)

    
    btn2 = Button(root, text="Simular", fg="black", font=("Cambria", 12), command =correr_simulacion)
    btn2["fg"]="black"
    btn2["bg"]="white"
    btn2["height"]=3
    btn2["width"]=15
    btn2["activebackground"]="#faf0ff"
    btn2.place(x=260,y=120)

    btn5 = Button(root, text="Resultados", fg="black", font=("Cambria", 12), command= w_resultados)
    btn5["fg"]="black"
    btn5["bg"]="white"
    btn5["height"]=3
    btn5["width"]=15
    btn5["activebackground"]="#faf0ff"
    btn5.place(x=445,y=120)
    

    
    root.mainloop()
    
```
